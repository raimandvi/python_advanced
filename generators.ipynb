## Generators in python

Generators are special functions that return values one at a time using the 'yield' keyword instead of returning all values at once.

**1. Generator**

concept: . return - returns all values.
         . yield - pauses and resumes.
        
         
          

  

```python
def num_generator(n):
  for i in range(n):
    yield i

for num in num_generator(4):
  print(num)
```

**2. Using next() with Generator**




```python
def count():
  yield 1
  yield 2
  yield 3

gen = count()
print(next(gen))
print(next(gen))
print(next(gen))
```

**3. Gnerator Expression**


Concept: Like list comprehension

```python
gen = (i * i for i in  range(1,10))

for value in gen:
  print(value)
```
