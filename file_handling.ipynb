## File handling in python

File handling allows us to create, read, write, and update files using python.

#File Modes


*   r : Read
*   w : write


*   a : append
*   x : create

*   rb : Read binary
*   wb : write binary







**1. Lambda function**
         
          

  

```python
square = lambda a: a * a
print (square(9))
```

**1. Reading from a file (read())**



```python
file = open("sample.txt", "r")
content = file.read()
print(content)
file.close()
```

**2. Writing to a file (write())**


*   Creates file if not exists
*   Overwrites content



```python
file = open("sample.txt", "w")
file.write("Hello, this is file handling in Python.")
file.close()
```

**3. Reading all lines (readlines())**



```python
file = open("sample.txt", "r")
lines = file.readlines()
print(lines)
file.close()
```

**4. Appending to a file (append mode))**








```python
file = open("sample.txt", "a")
file.write("\nThis line is appended.")
file.close()
```

**5. Using with statement**






```python
with open("sample.txt", "r") as file:
    data = file.read()
    print(data)
```

**6. File Handling with User Input**

```python
text = input("Enter text to write into file: ")

with open("user_input.txt", "w") as file:
    file.write(text)
```
