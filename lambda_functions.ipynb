## Lambda functions in python

Lambda functions are small anonymous functions defined using the 'lambda' keyword.                              They can take any number of arguments but can have only one expression.

**1. Lambda function**
         
          

  

```python
square = lambda a: a * a
print (square(9))
```

**2. Lambda with Multiple Arguments**




```python
addition = lambda a,b: a+b
print(addition(4, 5))
```

**3. Lambda with Conditional Expression**




```python
great_num = lambda x,y: x if x>y else y
print(great_num(9,10))
```

**4. Lambda with List**

 Often used with list

```python
numbers = [1, 2, 3, 4, 5]

cubes = list(map(lambda x: x**3, numbers))
print(cubes)
```

**5. Lambda with filter()**


* Filtering data






```python
numbers = [2,5,7,8,10,15,12]

even_num = list(filter(lambda x: x%2 == 0, numbers))
print(even_num)
```

**6.Lambda with reduce()**


*   Aggregation




```python
from functools import reduce

numbers = [1,2,3,4,5]

sum = reduce(lambda a, b: a+b, numbers)
print(sum)
```

**7. Lambda with user input**

```python
num = int(input("Enter a number: "))
pow = int(input("Enter the power: "))
res = lambda x, y: x**y
print("Power of given number: ",res(num, pow))
```
