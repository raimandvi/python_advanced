 Variable - Length Arguments (*args and **kwargs)


 *args and **kwargs are used when the number of arguments passed to a function is not known in advance.

**1. args (Variable Positional Argument)**

concept: . Accepts multiple values.
         . Stored as a tuple

```python
def pow_num(*args):
  power = int(input("Enter power: "))
  for num in args:
    print(num ** power)

pow_num(2,4,6,7)
```

**2. args with Normal Arguments**

```python
def stu_marks(name, *marks):
  print("Name:", name)
  print("Marks:", marks)

stu_marks("Ali", 20,40,70,95,70)
stu_marks("Payal", 75,85,59,95,99)
```

**3. kwargs (Variable Keyword Argument)**


Concept : Stored a s dictionary
          Key-value pairs

```python
def stu_info(**kwargs):
  for key, value in kwargs.items():
    print(key, ":", value)

stu_info(name="Ali", age=25, course="Data Science")
```

**4. kwargs with fixed parameters**

```python
def stu_tripinfo(Roll_no, **places):
  print("Roll no:", Roll_no)
  for key, value in places.items():
    print(key, ":", value)

stu_tripinfo("20", Place = "Mumbai", expenses = 4000, time = "4:30")
stu_tripinfo("25", Place = "Nanital", expenses = 5000, time = "2:30")



```

**5. args and kwargs together**

```python
def order_details(order_id, *items, **details):
    print("Order ID:", order_id)
    print("Items:", items)
    print("Details:", details)

order_details(
    101,
    "Laptop", "Mouse",
    price=55000,
    delivery="Yes"
)
```
